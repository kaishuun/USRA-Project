```r
library(tidyverse)
library("networkD3")
library("rbokeh")
```

```r
#Prepares data by calculating impressions (impressions = vid views * #times channel is recommended)

flow <- read_csv("Combined Probability.csv") %>% select(-`X1`) %>% filter(!is.na(`Next Category`))
flow <- flow %>% mutate(temp = 1) %>% group_by(`Current Video`) %>% mutate(Numvids = sum(temp))
totalvids <- flow %>% group_by(`Category`) %>% count(`Current Video`) %>% mutate(temp = 1) %>% summarize(totalvids = sum(temp))
flow <- flow %>% left_join(totalvids) %>% mutate(`Average Impressions` = `Current Views` * Amount /totalvids) %>% select(Category, `Next Category`, `Average Impressions`)
flow <- flow %>% group_by(Category,`Next Category`) %>% summarize(`Average Impressions` = sum(`Average Impressions`)) %>% ungroup()
flow <- flow %>% mutate(`Next Category` = paste(flow$`Next Category`,"_"))

#this one removes LateNightTalkShow
flow <- flow %>% filter((Category != "LateNightTalkShow"),(`Next Category` != "LateNightTalkShow _"))

flow
```

```r
nodes <- data.frame(names = unique(flow$"Category"),stringsAsFactors = FALSE)
nodes2 <- data.frame(names = unique(flow$"Next Category"),stringsAsFactors = FALSE)
nodes <- rbind(nodes,nodes2)
nodes <- nodes%>% unique()
rows <- nrow(nodes) -1
nodes <- cbind(nodes, node = c(0:rows))
```

```r
flow <- merge(flow,nodes,by.x = "Category", by.y = "names")
flow <- merge(flow,nodes,by.x = "Next Category", by.y = "names")
links <- flow[c("node.x","node.y","Average Impressions")]
colnames(links) <- c("source","target","value")
links
```

```r
sn <- sankeyNetwork(Links = links, Nodes = nodes, 
                         Source = 'source', 
                         Target = 'target', 
                         Value = 'value', 
                         NodeID = 'names',
                         units = 'Impressions')
saveNetwork(sn,"Scaled Channel Flow without LateNightTalkShow .html")
sn
```
