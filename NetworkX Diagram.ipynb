```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Community Recommendation Data')
df
```

```python
G = nx.DiGraph()
for i in range(len(df)):
    G.add_edge(df.iloc[i]['Origin Channel'],df.iloc[i]['Step: 1 Channel'],penwidth = 1)
    G.add_edge(df.iloc[i]['Step: 1 Channel'],df.iloc[i]['Step: 2 Channel'],penwidth = 1)
    G.add_edge(df.iloc[i]['Step: 2 Channel'],df.iloc[i]['Step: 3 Channel'],penwidth = 1)

    color_map = []
for node in G:
    if node in df['Origin Channel'].values:
        color_map.append("red")
    else:
        color_map.append("blue")
    
#nx.draw_networkx(G,node_color = color_map,with_label = True,layout=nx.spring_layout(G))
nx.draw(G, layout=nx.spring_layout(G),node_color = color_map,with_label = True)
plt.show()
```
